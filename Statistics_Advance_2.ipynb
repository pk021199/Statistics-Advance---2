1. What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?
 - Type I and Type II Errors in Hypothesis Testing
     
     In hypothesis testing, we start with two statements:
     * Null hypothesis (H₀): No effect or no difference.
     * Alternative hypothesis (H₁ or Ha): There is an effect or difference.
     
     Because decisions are based on sample data (not the entire population), mistakes can occur. These mistakes are called Type I and Type II errors.

     Type I Error (False Positive)

     Definition:
     
     A Type I error occurs when we reject the null hypothesis even though it is actually true.
     
     In simple words:
     
     We think something has happened, but actually nothing happened.
     
     Example:
     
     A medical test says a patient has a disease, but actually the patient is healthy.

     Probability of Type I Error:       α=P(Type I error)

     Impact on decision-making:
     * Too high α → more false positives.
     * Organizations might waste resources acting on an effect that doesn’t really exist.

     Type II Error (False Negative)

     Definition:
     
     A Type II error happens when we fail to reject the null hypothesis even though it is false.
     
     In simple words:
     
     We say nothing happened, but actually something did happen.
     
     Example:
     
     A medical test says a patient is healthy, but actually they have the disease.

     Probability of Type II Error:  β=P(Type II error)

     Impact on decision-making:
     * Too high β → miss important effects.
     * Serious consequences when safety is involved (e.g., failing to detect a defective product or disease).

     Relationship Between Type I and Type II Errors
     * Reducing Type I error (α) usually increases Type II error (β).
     * Reducing Type II error usually increases Type I error.
       
       You must balance them depending on the situation.



2. What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?
 -  The P-value is the probability of obtaining a test statistic (or sample result) as extreme as or more extreme than the one observed, assuming the null hypothesis (H₀) is true.

     In simple words:  The P-value tells us how likely our sample result is, if H₀ is actually true.

     Interpretation of P-value in the context of the null hypothesis
  * A small P-value (usually < 0.05) means:
    * he observed data is very unlikely assuming H₀ is true.
    * So, we have evidence against the null hypothesis.
    * Reject H₀
  * A large P-value (≥ 0.05) means:
    * The observed data is likely under the null hypothesis.
    * There is not enough evidence to reject H₀.
    * Fail to reject H₀

     Decision Rule
    * If P-value ≤ α → Reject H₀
    * If P-value > α → Fail to reject H₀

     Important Points
    * A low P-value does not prove the alternative hypothesis; it only suggests H₀ is unlikely.
    * A high P-value does not prove H₀ is true; it only means there is not enough evidence against it.
    * P-value measures evidence against the null hypothesis, not the probability that the null hypothesis is true.

3. Explain the difference between a Z-test and a T-test, including when to use
each.
 -  Use a Z-test when:
     * Population standard deviation (σ) is known, and
     * Sample size is large (n ≥ 30)

       (Central Limit Theorem ensures normality)
     
     Examples:
     * Testing population mean of a large sample
     * Quality control with known σ
     * Proportion tests (Z-test is used for proportions)

   * Use a T-test when:
     * Population standard deviation is unknown, and
     * Sample size is small (n < 30)
       
       (t-distribution accounts for variability)
     Examples:
     * Comparing average marks of two small groups
     * Medical trial with small sample
     * Before-after experiments with limited participants

     Why T-test for small samples?
     * The t-distribution has heavier tails → gives more conservative (safer) results.
     * It adjusts for more uncertainty when estimating σ from a small sample.

4. What is a confidence interval, and how does the margin of error influence
its width and interpretation?
 - A confidence interval (CI) is a range of values that is likely to contain the true population parameter (such as mean or proportion) with a certain level of confidence (usually 95%).

     In simple words:  A confidence interval tells us how much we can trust our sample estimate.

     Example (95% CI for mean):  CI=xˉ±margin of error

   What is the Margin of Error?
   
   The margin of error (ME) is the amount added and subtracted from the sample estimate to form the confidence interval.

  * How the Margin of Error Influences the Width of the CI
    1. Larger Margin of Error → Wider Confidence Interval
       * Indicates more uncertainty about the estimate.
       * Caused by:
          * Higher confidence level (90% → 95% → 99%)
          * Higher standard deviation (more variability)
          * Smaller sample size (n small)
    2. Smaller Margin of Error → Narrower Confidence Interval
       * Indicates more precise estimates
       * Caused by:
         * Lower confidence level
         * Lower variability
         * Larger sample size (n large)

     Interpretation Impact
      * Wide interval → Less precise, more uncertainty about the population parameter.
      * Narrow interval → More precise estimate, stronger evidence.

      Example:
      * Wide CI: (50, 90) → very uncertain
      * Narrow CI: (68, 72) → very precise estimate

5. Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?
 - ANOVA is a statistical test used to determine whether there are significant differences between the means of three or more groups.

   Why is it needed?
    * A t-test compares two means.
    * When you have 3 or more groups, running multiple t-tests increases the risk of Type I error (false positives).
    * ANOVA solves this problem by testing all group means simultaneously with one test.

     Purpose (simple statement):
     
     ANOVA checks whether at least one group mean is significantly different from the others.

     Hypotheses in ANOVA
     * Null hypothesis (H₀):
       All group means are equal
       (µ₁ = µ₂ = µ₃ = ...)
     * Alternative hypothesis (H₁):
       At least one group mean is different
       (Not all means are equal)

  * How ANOVA Extends Hypothesis Testing to More Than Two Groups
     * Instead of using multiple t-tests (which raise error rates), ANOVA uses the F-statistic, comparing:
       * Variability between groups
       * Variability within groups

     If the between-group variation is large compared to within-group variation, ANOVA concludes that group means differ.
     
     Thus, ANOVA allows hypothesis testing across 3, 4, 5, or even 10+ groups with a single statistical test.

   * Assumptions of ANOVA
     
     ANOVA is valid only if three key assumptions hold:
     1. Normality
        * The data in each group should be approximately normally distributed.
     2. Homogeneity of Variance (Equal variances)
        * All groups should have similar variances.

          (Also called homoscedasticity)
     3. Independence of Observations
        * Each observation must be independent of others.

          (No repeated measures unless special ANOVA types like repeated-measures ANOVA are used)


6. Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.

```python
import numpy as np
from math import erf, sqrt

# ---------------------------
# 1. Sample Data (example)
# ---------------------------
data = np.array([52, 55, 49, 50, 53, 54, 51, 56, 52, 50])

# ---------------------------
# 2. Population Parameters
# ---------------------------
mu0 = 50        # Hypothesized population mean
sigma = 5        # Known population standard deviation

# ---------------------------
# 3. Compute Z-Statistic
# ---------------------------
xbar = np.mean(data)     # sample mean
n = len(data)

z_value = (xbar - mu0) / (sigma / np.sqrt(n))

# ---------------------------
# 4. Calculate p-value (two-tailed)
# ---------------------------
def normal_cdf(x):
    return (1.0 + erf(x / sqrt(2))) / 2.0

p_value = 2 * (1 - normal_cdf(abs(z_value)))

# ---------------------------
# 5. Print Results
# ---------------------------
print("Sample Mean:", xbar)
print("Z-Statistic:", z_value)
print("P-value:", p_value)

# ---------------------------
# 6. Interpretation
# ---------------------------
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion: The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion: There is no significant difference between sample mean and population mean.")
```

7. Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using
NumPy and plot the histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate data from a binomial distribution
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of simulated observations

data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(0, n+2), edgecolor='black')
plt.title("Histogram of Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

8. Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate data from a non-normal distribution (Exponential)
population = np.random.exponential(scale=1, size=100000)

# Step 2: Draw multiple samples and compute sample means
sample_size = 50      # n
num_samples = 1000    # Number of samples

sample_means = []

for i in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# Step 3: Plot the original (non-normal) distribution
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.hist(population, bins=50, edgecolor='black')
plt.title("Original Non-normal Distribution (Exponential)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Step 4: Plot the distribution of sample means
plt.subplot(1,2,2)
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("Distribution of Sample Means (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

9. Write a Python function to calculate and visualize the confidence interval
for a sample mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def confidence_interval_plot(data, confidence=0.95):
    """
    Calculates and visualizes the confidence interval for the sample mean.

    Parameters:
        data (array-like): Sample data
        confidence (float): Confidence level (default = 0.95)
    """

    # Sample statistics
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)

    # Standard Error
    se = std / np.sqrt(n)

    # t-critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Confidence interval
    lower = mean - t_crit * se
    upper = mean + t_crit * se

    print(f"Sample Mean: {mean:.4f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.4f}, {upper:.4f})")

    # Visualization
    plt.figure(figsize=(8, 5))
    plt.axvline(mean, color='blue', linestyle='--', label='Sample Mean')
    plt.axvline(lower, color='red', linestyle='--', label='Lower CI Limit')
    plt.axvline(upper, color='green', linestyle='--', label='Upper CI Limit')

    plt.hist(data, bins=20, alpha=0.5, edgecolor='black')
    plt.title(f"Confidence Interval for Sample Mean ({int(confidence*100)}%)")
    plt.xlabel("Data Values")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()

    return (lower, upper)

# Example usage:
# data = np.random.normal(50, 10, 100)
# confidence_interval_plot(data)
```

10.  Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example)
observed = np.array([8, 12, 9, 11, 10, 10])

# Expected frequencies
expected = np.array([10, 10, 10, 10, 10, 10])

# Chi-Square Goodness-of-Fit Test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("p-value:", p_value)

# Interpretation
alpha = 0.05
if p_value > alpha:
    print("Fail to reject H0: Observed distribution fits the expected distribution.")
else:
    print("Reject H0: Observed distribution does NOT fit the expected distribution.")
```
